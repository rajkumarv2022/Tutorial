```python
data=pd.read_csv("Mall_Customers.csv")
```

```python
data.head()
```

```python
X=data[['Annual Income (k$)','Spending Score (1-100)']]
```

```python
from sklearn.cluster import KMeans
knn=KMeans(n_clusters=5)
knn.fit(X)
```

```python
knn.predict(X)
```

```python
y=knn.predict(X)
X['Cluster']=y
X.head()
```

```python
import seaborn as sns
sns.scatterplot(data=X,x='Annual Income (k$)',y='Spending Score (1-100)',hue='Cluster',palette=['red','green','blue','yellow','orange'])
```
